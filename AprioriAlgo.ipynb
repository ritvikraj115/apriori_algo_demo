```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df= pd.read_csv(r'app.csv')
```

```python
df.head()
```

```python
df['Product Description']=df['Product Description'].str.split(',')
```

```python
df
```

```python
inv=[]
pr=[]
count=0
for j in df['Product Description']:
    count=count+1
    for k in range(len(j)):
            inv.append(count)
            pr.append(j[k])
            data={'InvoiceNo':inv,'product':pr,'Quantity':1}
      
data=pd.DataFrame(data)
```

```python
data
```

```python
basket = (data.groupby(['InvoiceNo', 'product'])['Quantity']
.sum().unstack().reset_index().fillna(0) 
.set_index('InvoiceNo'))
```

```python
basket
```

```python
frequent_itemsets = apriori (basket, min_support=0.07, use_colnames=True)
rules = association_rules (frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
def pred(x):
    x=frozenset({x})
    p=rules[rules['antecedents']==x]
    print(p)
    p=p.sort_values(by='confidence',ascending=False)
    print(p)
    
```

```python
pred('Bread')
```
